## Creating DataFrame from dictionary

```python
df = {'col_1' : [3,2,1,0], 'col_2' : ['a', 'b', 'c', 'd']}
```

```python
df0 = pd.DataFrame(df)
df0
```

Specify orient='index' to create the DataFrame using dictionary keys as rows:

```python
df1 = pd.DataFrame.from_dict(df, orient='index')
df1
```

When using the ‘index’ orientation, the column names can be specified manually:

```python
df2 = pd.DataFrame.from_dict(df, orient='index', columns = ['A','B','C','D'])
df2
```

We can change the columns' name

```python
df2.columns
```

```python
df2.columns = ['first', 'second', 'third', 'fourth']
```

```python
df2
```

```python
df2.columns.values[0] = 'zero'
```

```python
df2
```

```python
df2.columns
```

```python
col = {'zero' : 'satu'}
df2.rename(columns = col, inplace = True)
```

```python
df2
```
